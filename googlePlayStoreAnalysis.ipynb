# Cleaning Data to perform operation

### Removing missing data

```python
#Getting missing data
totalmissing=data.isnull().sum()
totalmissing
```

```python
#removing missing data
data=data.dropna(how='any')
#getting missing data
data.isnull().sum()
```

### Cleaning Size Columns

```python
#looking and working on size
pd.unique(data.Size)
```

```python
#getting number of app having size = varies with device
len(data[data.Size=='Varies with device'])
```

```python
# Converting varies with device to NA
data['Size'].replace('Varies with device',np.nan,inplace=True)
pd.unique(data.Size)
```

```python
#removing K and M from last of size and converting to float
a=data.Size.replace(r'[kM]+$','',regex=True).astype(float) 
#converting K to 10**3 and M to 10**6
b=data.Size.str.extract(r'[\d\.]+([kM]+)',expand=False).fillna(1).replace(['k','M'],[10**3,10**6]).astype(int) 
```

```python
#get size in bytes
data.Size=a*b 
```

```python
#filling all NaN value with mean
data['Size'].fillna(data.groupby('Category')['Size'].transform('mean'),inplace = True)
```

```python
pd.unique(data.Size)# after removing K and M
```

### Cleaning Installs Columns

```python
pd.unique(data.Installs)
```

```python
#Removing + and convertin from object to int data type
data.Installs=data.Installs.replace(r'[,+]','',regex=True).astype(int)
pd.unique(data.Installs)
```

```python
sorted_data=sorted(list(data['Installs'].unique()))
data['Installforplot']=data['Installs'].replace(sorted_data,range(0,len(sorted_data),1))
```

### Cleaning Price Columns

```python
data.Price.value_counts().head(20)
```

```python
#Removing doller sign
data.Price=data.Price.replace(r'[,$]','',regex=True).astype(float)
```

```python
#Last Updation
data['Last Updated']=pd.to_datetime(data['Last Updated'])
```
